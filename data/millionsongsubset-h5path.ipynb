```python
import pandas as pd
import os
from os import listdir
```

```python
basepath = 'MillionSongSubset/'
h5path = []
```

```python
layer1 = listdir(basepath)
for i, folder1 in enumerate(layer1):
    if folder1 == '.DS_Store':
        os.remove(basepath + folder1)
    else:
        basepath1 = basepath + folder1
        #print(basepath1)
        layer2 = listdir(basepath1)
        
    for j, folder2 in enumerate(layer2):
        #print(folder2)
        if folder2 == '.DS_Store':
            os.remove(basepath1 + '/' + folder2)
        else:
            basepath2 = basepath1 + '/' + folder2
            #print(basepath2)
            layer3 = listdir(basepath2)

        for folder3 in layer3:
            if folder3 == '.DS_Store':
                os.remove(basepath2 + '/' + folder3)
            else:
                basepath3 = basepath2 + '/' + folder3
                #print(basepath3)
                #h5path.append(basepath2)
                layer4 = listdir(basepath3)
            
            for folder4 in layer4:
                if folder4 == '.DS_Store':
                    os.remove(basepath3 + '/' + folder4)
                else:
                    basepath4 = basepath3 + '/' + folder4
                    h5path.append(basepath4)
```

```python
# save path files to txt
with open('h5path.txt', mode='wt', encoding='utf-8') as myfile:
    myfile.write('\n'.join(h5path))
```
